```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob

import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
survfiles = sorted(glob.glob('100turns/surv*.npy'))
qxsfiles = sorted(glob.glob('100turns/qxtuned*.npy'))
x_stdfiles = sorted(glob.glob('100turns/x_std*.npy'))

survs = np.array([])
qxs = np.array([])
sigmax_growths = np.array([])

for survfilei,qxfilei,x_stdfilei in zip(survfiles,qxsfiles,x_stdfiles):
    survi = np.load(survfilei)
    survs = np.append(survi,survs)

    qxi = np.load(qxfilei)
    qxs = np.append(qxi,qxs)

    x_stdi = np.load(x_stdfilei)
    sigmax_growthi = np.max(x_stdi)/x_stdi[0]
    sigmax_growths = np.append(sigmax_growthi,sigmax_growths)
    
```

```python
fig,ax = plt.subplots(1,1,figsize = (11,8))

# fig.set_dpi(150)

ax.plot(qxs,survs, marker = 'x', linewidth = 3,markersize = 10,linestyle = '--')

ax.set_xlabel(r'$Q_x$',fontsize =50)
ax.set_ylabel(r'Beam Survival Ratio',fontsize =42)

ax.tick_params(axis='both', labelsize=40)

axt = ax.twinx()

axt.plot(qxs,sigmax_growths, c = 'orange', marker = '^',linestyle = '--',linewidth = 2,markersize = 10)

# axt.set_ylabel(r'$\sigma_{x,MAX} / \sigma_{x,0}$',fontsize =42)
axt.set_ylabel(r'Max. Beam Size Growth',fontsize =42)

axt.tick_params(axis='both', labelsize=40)

plt.show()
plt.close()
```
